```python
import cv2
import matplotlib.pyplot as plt
```

```python
from skimage.feature import hog
from skimage import data,exposure
```

```python
from skimage.feature import hog
```

```python
img = cv2.imread("leo.jpg")
img
```

```python
plt.imshow(img)
```

```python
featdis, hog_image = hog(img,pixels_per_cell=(8, 8), cells_per_block=(1,1), visualize=True, channel_axis=2)
```

```python
print(featdis)
```

```python
cv2.imshow("hog",hog_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Rescaling

```python
rescale = exposure.rescale_intensity(hog_image, in_range =(0,1)) 
rescale
```

## Plot

```python
figure,(a1, a2) = plt.subplots(1, 2 ,figsize=(12, 6), sharex=True, sharey=True)

a1.axis('off')
a1.imshow(img)
a1.set_title('Input image')

a2.axis('off')
a2.imshow(rescale)
a2.set_title('Histogram of Oriented Greadiets')
plt.show()
```
